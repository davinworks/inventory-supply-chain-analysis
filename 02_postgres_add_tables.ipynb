```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:davinworks123@localhost:5432/inventory_supply_chain')

# Import distribution_centers
df = pd.read_csv(r'E:\Inventory and Supply Chain Analyst\distribution_centers.csv')
df = df[['id', 'name', 'latitude', 'longitude']]
df.to_sql('distribution_centers', engine, if_exists='replace', index=False)
print("distribution_centers: OK", df.shape)
```

```python
# Import products
df_products = pd.read_csv(r'E:\Inventory and Supply Chain Analyst\products.csv')
df_products.to_sql('products', engine, if_exists='replace', index=False)
print("products: OK", df_products.shape)

# Import orders
df_orders = pd.read_csv(r'E:\Inventory and Supply Chain Analyst\orders.csv')
df_orders.to_sql('orders', engine, if_exists='replace', index=False)
print("orders: OK", df_orders.shape)

# Import order_items
df_order_items = pd.read_csv(r'E:\Inventory and Supply Chain Analyst\order_item.csv')
df_order_items.to_sql('order_items', engine, if_exists='replace', index=False)
print("order_items: OK", df_order_items.shape)

# Import inventory_items
df_inventory = pd.read_csv(r'E:\Inventory and Supply Chain Analyst\inventory_items.csv')
df_inventory.to_sql('inventory_items', engine, if_exists='replace', index=False)
print("inventory_items: OK", df_inventory.shape)
```
